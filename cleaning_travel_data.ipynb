```python
#Clean airport travel data

#import dependencies
import pandas as pd
```

```python
#cleaning objectives
#remove unnecessary data thats outside out date range. (jan 1st 2011 - dec 31st 2021)
#remove "total" row from dfs
#add a city column
#consolidate airport data to one df per city
#change data type for Month and TOTAL to int-64

#manipulation objectives
#consolidate all monthly data for each city into one df 
#consolidate all annual data for each city into one df
```

```python
#load in travel data
boston_csv = "Clean_data/boston.csv"
chi_midway_csv = "Clean_data/Chicago Midway.csv"
chi_ohare_csv = "Clean_data/Chicago O'Hare.csv"
las_vegas_csv = "Clean_data/Las Vegas.csv"
la_csv = "Clean_data/LAX new.csv"
miami_csv = "Clean_data/Miami.csv"
ny_jfk_csv = "Clean_data/NY JFK.csv"
ny_guardia_csv= "Clean_data/NY La Guardia.csv"
orlando_csv = "Clean_data/Orlando.csv"
san_fran_csv = "Clean_data/San Francisco.csv"
seattle_csv = "Clean_data/Seattle.csv"
dc_dulles_csv = "Clean_data/Washington DC Dulles.csv"
dc_ronald_csv = "Clean_data/Washington DC Ronald Reagen.csv"

```

```python
#Boston
#create df for each city
b = pd.read_csv(boston_csv)

boston_df = b.loc[(b["Year"] > 2010) & (b["Year"] < 2022), :]

boston_df["City"] = "Boston"

#Remove commas and change data type for total column 
boston_df['TOTAL'] = boston_df['TOTAL'].str.replace(',','')
boston_df.astype({'TOTAL': 'int64'})

boston_df.head()
```

```python
#create df for each city
c = pd.read_csv(chi_midway_csv)

c_ohare = pd.read_csv(chi_ohare_csv)

#add city column
c["City"] = "Chicago"
c_ohare["City"] = "Chicago"

#consolidate data into one df
chicago_df = pd.merge(c_ohare, c, how="outer")

chicago_df = chicago_df.loc[(chicago_df["Year"] > 2011) & (chicago_df["Year"] < 2022), :]

#Remove commas and change data type for total column 
chicago_df['TOTAL'] = chicago_df['TOTAL'].str.replace(',','')
chicago_df.astype({'TOTAL': 'int64'})

#chicago_df.head()
```

```python
#las vegas
#create df for each city
lv = pd.read_csv(las_vegas_csv)

vegas_df = lv.loc[(lv["Year"] > 2011) & (lv["Year"] < 2022), :]

vegas_df["City"] = "Las Vegas"

#Remove commas and change data type for total column 
vegas_df['TOTAL'] = vegas_df['TOTAL'].str.replace(',','')
vegas_df.astype({'TOTAL': 'int64'})

vegas_df.head()
```

```python
#miami
#create df for each city
m = pd.read_csv(miami_csv)

miami_df = m.loc[(m["Year"] > 2011) & (m["Year"] < 2022), :]

miami_df["City"] = "Miami"

#Remove commas and change data type for total column 
miami_df['TOTAL'] = miami_df['TOTAL'].str.replace(',','')
miami_df.astype({'TOTAL': 'int64'})

miami_df.head()
```

```python
#LA
#create df for each city
la = pd.read_csv(la_csv)

los_angeles_df = la.loc[(la["Year"] > 2011) & (la["Year"] < 2022), :]

los_angeles_df["City"] = "Los Angeles"

#Remove commas and change data type for total column 
los_angeles_df['TOTAL'] = los_angeles_df['TOTAL'].str.replace(',','')
los_angeles_df.astype({'TOTAL': 'int64'})

los_angeles_df.head()
```

```python
#new york 
#create df for each city
ny = pd.read_csv(ny_guardia_csv)

ny_jfk = pd.read_csv(ny_jfk_csv)

#add city column
ny["City"] = "New York"
ny_jfk["City"] = "New York"

#consolidate data into one df
ny_df = pd.merge(ny, ny_jfk, how="outer")

ny_df = ny_df.loc[(ny_df["Year"] > 2011) & (ny_df["Year"] < 2022), :]

#Remove commas and change data type for total column 
ny_df['TOTAL'] = ny_df['TOTAL'].str.replace(',','')
ny_df.astype({'TOTAL': 'int64'})

ny_df.head()
```

```python
#Orlando
#create df for each city
o = pd.read_csv(orlando_csv)

orlando_df = o.loc[(o["Year"] > 2011) & (o["Year"] < 2022), :]

orlando_df["City"] = "Orlando"

#Remove commas and change data type for total column 
orlando_df['TOTAL'] = orlando_df['TOTAL'].str.replace(',','')
orlando_df.astype({'TOTAL': 'int64'})

orlando_df.head()
```

```python
#San Fransico
#create df for each city
sf = pd.read_csv(san_fran_csv)

san_fran_df = sf.loc[(sf["Year"] > 2011) & (sf["Year"] < 2022), :]

san_fran_df["City"] = "San Francisco"

#Remove commas and change data type for total column 
san_fran_df['TOTAL'] = vegas_df['TOTAL'].str.replace(',','')
san_fran_df.astype({'TOTAL': 'int64'})

san_fran_df.head()
```

```python
#seattle
#create df for each city
s = pd.read_csv(seattle_csv)

seattle_df = s.loc[(s["Year"] > 2011) & (s["Year"] < 2022), :]

seattle_df["City"] = "Seattle"

#Remove commas and change data type for total column 
seattle_df['TOTAL'] = seattle_df['TOTAL'].str.replace(',','')
seattle_df.astype({'TOTAL': 'int64'})

seattle_df.head()
```

```python
#Washington DC
#create df for each city
dc = pd.read_csv(dc_dulles_csv)

dc_ronald = pd.read_csv(dc_ronald_csv)

#add city column
dc["City"] = "Washington DC"
dc_ronald["City"] = "Washington DC"

#consolidate data into one df
dc_df = pd.merge(dc, dc_ronald, how="outer")

dc_df = dc_df.loc[(dc_df["Year"] > 2011) & (dc_df["Year"] < 2022), :]

#Remove commas and change data type for total column 
dc_df['TOTAL'] = dc_df['TOTAL'].str.replace(',','')
dc_df.astype({'TOTAL': 'int64'})

dc_df.head()
```

```python
city_list = [chicago_df, los_angeles_df, vegas_df, ny_df, orlando_df, san_fran_df, seattle_df, dc_df]
total_arrivals = boston_df.loc[(boston_df["Month"] == "TOTAL"), :]
monthly_arrivals = boston_df[boston_df["Month"] != "TOTAL"]

#Create a Dataframe for annual results for all 10 cities
for city in city_list:
    temp_city = city.loc[(city["Month"] == "TOTAL"), :]
    total_arrivals = pd.merge(total_arrivals, temp_city, how="outer")

#Create a Dataframe for monthly results for all 10 cities
for place in city_list:
    temp_place = place[place["Month"] != "TOTAL"]
    monthly_arrivals = pd.merge(monthly_arrivals, temp_place, how="outer")

#change data type of month to integer now that the "TOTAL" value has been removed
monthly_arrivals = monthly_arrivals.astype({'Month': 'int64'})
monthly_arrivals = monthly_arrivals.astype({'TOTAL': 'int64'})

total_arrivals = total_arrivals.astype({'TOTAL': 'int64'})

```

```python
import psycopg2
from sqlalchemy import create_engine, func
import config
```

```python
# Connect to your Postgres database
engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(config.username, config.password, 'localhost', 5432, 'city_weather_db'))
con = engine.connect()
```

```python
# use pandas to load monthly_arrivals dataframe into the database
monthly_arrivals.to_sql('monthly_arrivals', con, if_exists='replace', index=False)

# use pandas to load total_arrivals dataframe into the database
total_arrivals.to_sql('total_arrivals', con, if_exists='replace', index=False)
```
