```python
weather_data["AIRPORT"] = weather_data["NAME"].str.split(",")
weather_data.head()
```

```python
#Rename 'Name' column to 'Airport Code'

weather_data.rename(columns={"NAME": "AIRPORT CODE"}, inplace=True)
weather_data.head()
```

```python
#Rearranging Columns 

weather_data = weather_data[["AIRPORT CODE", "AIRPORT", "LATITUDE", "LONGITUDE", "ELEVATION", "DATE",
                            "AWND", "WSF5", "PRCP", "SNOW", "TAVG", "TMAX", "TMIN"]]
weather_data.head()
```

```python
#Chaning Column Names

weather_data.rename(columns={"AWND":"Average Wind Speed", "WSF5":"Max Gust Speed",
                                "PRCP":"Precipitation", "SNOW": "Snow Accumulation",
                                "TAVG":"Average Temperature", "TMAX":"Max Temperature",
                                 "TMIN":"Min Temperature"}, inplace=True)
weather_data.head()
```

```python
#Dataframe showing averages of columns

monthly_summary = weather_data.groupby(["DATE", "AIRPORT CODE", "LATITUDE", "LONGITUDE", "ELEVATION"]).mean()
monthly_summary.head()
```
